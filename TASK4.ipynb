SAVING THE VALUES INTO FILES BASED ON CODEKATA SCORE --EXCEEDED EXPECTATIONS, REACHED EXPECTATIONS, NEEDS IMPROVEMENT AND UNSATISFACTORY

```python
for i in range(len(ls)):
  if (ls[i]>15000):
    dfee=res.iloc[i]
    dfee.to_csv('Exceeded_expectations.csv', mode='a', index=False, header=False)
  elif ((ls[i]>10000) and (ls[i]<15000)):
    dfre=res.iloc[i]
    dfre.to_csv('Reached_expectations.csv', mode='a', index=False, header=False)
  elif ((ls[i]>7000) and (ls[i]<10000)):
    dfni=res.iloc[i]
    dfni.to_csv('Needs_improvement.csv', mode='a', index=False, header=False)
  elif (ls[i]<7000):
    dfu=res.iloc[i]
    dfu.to_csv('Unsatisfactory.csv', mode='a', index=False, header=False)
  else:
    print("0")
```
